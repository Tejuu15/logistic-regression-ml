```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df = pd.read_csv("Titanic-Dataset.csv")

# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Convert categorical to numeric
df['Sex'] = df['Sex'].map({'male':1, 'female':0})

# Select important features
df = df[['Age', 'Pclass', 'Sex', 'Fare', 'Survived']]
```

```python
X = df[['Age', 'Pclass', 'Sex', 'Fare']]
y = df['Survived']
```

```python
# Display the first 5 rows of the features (X)
display(X.head())
```

```python
# Display the first 5 rows of the target variable (y)
display(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

These shapes confirm that your data has been successfully split into training and testing sets. `X_train` and `y_train` contain 80% of your data, and `X_test` and `y_test` contain the remaining 20%.

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

After scaling the features, the next step is to initialize and train a logistic regression model. This model will learn the relationship between the features (`X_train`) and the target variable (`y_train`) to make predictions on unseen data.

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
```

Now that we have the predictions, let's evaluate the model's performance using various metrics such as accuracy, precision, recall, and ROC AUC score.

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_prob)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")
```
